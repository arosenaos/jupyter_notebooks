## multivariate self organizing maps project description

We can use Z500 and IVT to train SOM over SGP, and identify leading very dry and very wet nodes based on composite precipitation for each node, and evaluate composite soil moisture anomalies for these very dry and wet nodes.  Also, evaluate the correlation between each of the very wet and very dry nodes with AMV, ENSO, NAO and PNA and PDV indices.

### read in precipitation 

```python
precip_dir = "/home/annierosen16/data/texas_soms/precip"

files = sorted(glob.glob(os.path.join(precip_dir, "precip_*.nc")))

print(f"Found {len(files)} files")

precip = xr.open_mfdataset(files, combine='by_coords', engine='netcdf4',
                            chunks={'valid_time': 500})

tx_ok_lat_min, tx_ok_lat_max = 26.0, 37.0

tx_ok_lon_min, tx_ok_lon_max = -106.0, -94.0

precip = precip.sel(
    
    latitude=slice(tx_ok_lat_max, tx_ok_lat_min),
    
    longitude=slice(tx_ok_lon_min, tx_ok_lon_max))

print(f"Precip lats: {precip.latitude.values[[0,-1]]}")

print(f"Precip lons: {precip.longitude.values[[0,-1]]}")

print(f"Precip shape: {precip['tp'].shape}")
```

```python
# 1. daily total precip in mm — min_count=4 ensures all 4 six-hourly timesteps present

precip_daily = precip['tp'].resample(valid_time='1D').sum(min_count=4) * 1000

# 2. drop partial days (fewer than 4 timesteps)

precip_daily = precip_daily.dropna(dim='valid_time', how='any')

# 3. warm season only (Apr-Sep)

precip_daily = precip_daily.sel(valid_time=precip_daily.valid_time.dt.month.isin([4, 5, 6, 7, 8, 9]))

# 4. compute into memory

print("Computing precip_daily...")

precip_daily = precip_daily.compute()

print(f"precip_daily shape: {precip_daily.shape}")

print(f"Months present: {np.unique(precip_daily.valid_time.dt.month.values)}")

print(f"Raw precip range (mm): {float(precip_daily.min()):.4f} to {float(precip_daily.max()):.4f}")
```

```python
# 5. climatology

precip_clim_mean = precip_daily.groupby('valid_time.dayofyear').mean('valid_time')

precip_clim_std  = precip_daily.groupby('valid_time.dayofyear').std('valid_time')

# 6. doy index mapping

clim_doys_precip  = precip_clim_mean.dayofyear.values

doy_to_idx_precip = {doy: i for i, doy in enumerate(clim_doys_precip)}

doy_precip        = precip_daily.valid_time.dt.dayofyear.values

precip_idx        = np.array([doy_to_idx_precip[d] for d in doy_precip])

# 7. extract numpy arrays

precip_mean_np = precip_clim_mean.values

precip_std_np  = precip_clim_std.values

precip_np      = precip_daily.values

# 8. standardized anomaly

precip_anom_np = (precip_np - precip_mean_np[precip_idx]) / precip_std_np[precip_idx]

precip_anom_np = np.nan_to_num(precip_anom_np, nan=0.0)

print(f"precip anomaly range: {precip_anom_np.min():.2f} to {precip_anom_np.max():.2f}")

print(f"precip anomaly shape: {precip_anom_np.shape}")

# 9. spatial mean over Texas — one value per day

precip_anom_texas = precip_anom_np.mean(axis=(1, 2))

print(f"Texas mean precip anom shape: {precip_anom_texas.shape}")

print(f"Texas mean precip anom range: {precip_anom_texas.min():.2f} to {precip_anom_texas.max():.2f}")

# 10. retain dates alongside the daily texas mean anomaly

precip_dates = precip_daily.valid_time.values

print(f"Precip dates shape: {precip_dates.shape}")

print(f"Precip dates range: {precip_dates[0]} to {precip_dates[-1]}")
```

```python
path = '/data/rong4/yzhuang/Data/ERA5_1deg/geopotential'

files = sorted(glob.glob(f'{path}/era5.geopotential.*.nc'))

def in_range(f):
    
    yyyymm = int(os.path.basename(f).split('.')[2])
    
    return 197901 <= yyyymm <= 202512

def is_warm_season(f):
    
    month = int(os.path.basename(f).split('.')[2][-2:])
    
    return month in [4, 5, 6, 7, 8, 9]

files = [f for f in files if in_range(f) and is_warm_season(f)]

# load old geo

geo = xr.open_mfdataset(
    
    files,
    
    combine='by_coords',
    
    chunks={'time': 500, 'latitude': 51, 'longitude': 101})

# subset to CONUS domain

lon_min, lon_max = 234.0, 294.0

lat_min, lat_max =  24.0,  50.0

geo = geo.sel(
    
    longitude=slice(lon_min, lon_max),
    
    latitude=slice(lat_max, lat_min))

# match coordinate dtypes

geo = geo.assign_coords(
    
    longitude=geo.longitude.values.astype('float64'),
    
    latitude=geo.latitude.values.astype('float64')
)

print(f"Old geo lons: {geo.longitude.values[[0,-1]]}")

print(f"Old geo lats: {geo.latitude.values[[0,-1]]}")

print(f"Old geo time range: {geo.time.values[0]} to {geo.time.values[-1]}")

print(f"Old geo shape: {geo['z'].shape}")

# load new geo (2024-2025)

new_geo_dir = "/home/annierosen16/data/texas_soms/500_geo_2024_2025"

new_files = sorted(glob.glob(os.path.join(new_geo_dir, "geo500_*.nc")))

print(f"\nFound {len(new_files)} new geo files")

geo_new = xr.open_mfdataset(
    
    new_files,
    
    combine='by_coords',
    
    chunks={'valid_time': 500})

# clean new geo to match old

geo_new = geo_new.rename({'valid_time': 'time'})

geo_new = geo_new.squeeze('pressure_level').drop_vars('pressure_level')

geo_new = geo_new.drop_vars(['number', 'expver'], errors='ignore')

geo_new = geo_new.assign_coords(
    
    longitude=(geo_new.longitude.values % 360)).sortby('longitude')

geo_new = geo_new.assign_coords(
    
    longitude=geo_new.longitude.values.astype('float64'),
    
    latitude=geo_new.latitude.values.astype('float64'))

geo_new['z'] = geo_new['z'].astype('float64')

print(f"New geo lons: {geo_new.longitude.values[[0,-1]]}")

print(f"New geo lats: {geo_new.latitude.values[[0,-1]]}")

print(f"New geo time range: {geo_new.time.values[0]} to {geo_new.time.values[-1]}")

print(f"New geo shape: {geo_new['z'].shape}")

# verify coordinates match before concat

print(f"\nLons match: {np.allclose(geo.longitude.values, geo_new.longitude.values)}")

print(f"Lats match: {np.allclose(geo.latitude.values, geo_new.latitude.values)}")

# concat and compute z_height

geo_combined = xr.concat([geo, geo_new], dim='time').sortby('time')

geo_combined['z_height'] = geo_combined['z'] / 9.81

print(f"\nCombined time range: {geo_combined.time.values[0]} to {geo_combined.time.values[-1]}")

print(f"Combined shape: {geo_combined['z'].shape}")

print(f"Total timesteps: {len(geo_combined.time)}")

# NaN check

nan_count = geo_combined['z_height'].isnull().sum().compute()

print(f"\nNaN count: {int(nan_count)}")

print(f"NaN fraction: {float(nan_count) / geo_combined['z_height'].size:.4f}")
```

### ensure they contain the same time range

```python
t_start = max(geo_combined.time.values[0], ivt_combined.time.values[0])

t_end   = min(geo_combined.time.values[-1], ivt_combined.time.values[-1])

geo_common = geo_combined.sel(time=slice(t_start, t_end))

ivt_common = ivt_combined.sel(time=slice(t_start, t_end))

print(f"Common time range: {t_start} to {t_end}")

print(f"geo_common shape: {geo_common['z_height'].shape}")

print(f"ivt_common shape: {ivt_common['ivt'].shape}")
```

```python
geo_daily = geo_common['z_height'].chunk({'time': 500, 'latitude': 51, 'longitude': 101}).resample(time='1D').mean().astype('float32')

ivt_daily = ivt_common['ivt'].chunk({'time': 500, 'latitude': 51, 'longitude': 101}).resample(time='1D').mean().astype('float32')

print("Computing geo_daily...")

geo_daily = geo_daily.compute()

print("Computing ivt_daily...")

ivt_daily = ivt_daily.compute()

# filter to warm season after resampling

geo_daily = geo_daily.sel(time=geo_daily.time.dt.month.isin([4, 5, 6, 7, 8, 9]))

ivt_daily = ivt_daily.sel(time=ivt_daily.time.dt.month.isin([4, 5, 6, 7, 8, 9]))

print(f"geo_daily shape: {geo_daily.shape}")

print(f"ivt_daily shape: {ivt_daily.shape}")

print(f"Months present: {np.unique(geo_daily.time.dt.month.values)}")
```

```python
geo_clim_mean = geo_daily.groupby('time.dayofyear').mean('time')

geo_clim_std  = geo_daily.groupby('time.dayofyear').std('time')

ivt_clim_mean = ivt_daily.groupby('time.dayofyear').mean('time')

ivt_clim_std  = ivt_daily.groupby('time.dayofyear').std('time')

print("Climatology complete")
```

```python
# get the actual dayofyear values in the climatology

clim_doys = geo_clim_mean.dayofyear.values

doy_to_idx = {doy: i for i, doy in enumerate(clim_doys)}

# map each day's doy to its index in the climatology

doy_geo = geo_daily.time.dt.dayofyear.values

doy_ivt = ivt_daily.time.dt.dayofyear.values

geo_idx = np.array([doy_to_idx[d] for d in doy_geo])

ivt_idx = np.array([doy_to_idx[d] for d in doy_ivt])

# extract numpy arrays

geo_mean_np = geo_clim_mean.values

geo_std_np  = geo_clim_std.values

geo_np      = geo_daily.values

ivt_mean_np = ivt_clim_mean.values

ivt_std_np  = ivt_clim_std.values

ivt_np      = ivt_daily.values

# standardized anomalies

geo_anom_np = (geo_np - geo_mean_np[geo_idx]) / geo_std_np[geo_idx]

ivt_anom_np = (ivt_np - ivt_mean_np[ivt_idx]) / ivt_std_np[ivt_idx]

print("anomalies complete")

# replace NaNs from zero std divisions

geo_anom_np = np.nan_to_num(geo_anom_np, nan=0.0)

ivt_anom_np = np.nan_to_num(ivt_anom_np, nan=0.0)

print(f"geo anomaly range: {geo_anom_np.min():.2f} to {geo_anom_np.max():.2f}")

print(f"ivt anomaly range: {ivt_anom_np.min():.2f} to {ivt_anom_np.max():.2f}")

# reshape and concatenate

geo_vals = geo_anom_np.reshape(len(geo_daily.time), -1)

ivt_vals = ivt_anom_np.reshape(len(ivt_daily.time), -1)

X = np.concatenate([geo_vals, ivt_vals], axis=1)

print(f"SOM input shape: {X.shape}")
```

```python
# PCA

np.random.seed(42)

pca = PCA(n_components=55, random_state=42)

X_pca = pca.fit_transform(X)

cumvar = pca.explained_variance_ratio_.cumsum()

print(f"Variance explained by PCs: {cumvar[-1]:.2%}")

print(f"Reduced input shape: {X_pca.shape}")
```

```python
n_days = X_pca.shape[0]

som_size = 5

print(f"SOM grid size: {som_size}x{som_size}")

som = MiniSom(som_size, som_size, X_pca.shape[1],
              
              sigma=1.5,
              
              learning_rate=0.5,
              
              neighborhood_function='gaussian',
              
              random_seed=42)

som.pca_weights_init(X_pca)

print("Training SOM...")

som.train(X_pca, num_iteration=10000, verbose=True)

print("Done.")
```

```python
winners = np.array([som.winner(x) for x in X_pca])

node_labels = winners[:, 0] * som_size + winners[:, 1]

counts = np.bincount(node_labels, minlength=som_size**2)

print(f"Min days per node: {counts.min()}")

print(f"Max days per node: {counts.max()}")

print(f"Mean days per node: {counts.mean():.1f}")

print(f"Empty nodes: {(counts == 0).sum()}/{som_size**2}")
```

```python
som_weights = som.get_weights()

n_components = pca.n_components_

weights_2d = som_weights.reshape(-1, n_components)

weights_original = pca.inverse_transform(weights_2d)

n_lat = geo_daily.latitude.size   # 27

n_lon = geo_daily.longitude.size  # 61

n_spatial = n_lat * n_lon

geo_weights = weights_original[:, :n_spatial].reshape(som_size, som_size, n_lat, n_lon)

ivt_weights = weights_original[:, n_spatial:].reshape(som_size, som_size, n_lat, n_lon)

print(f"geo_weights shape: {geo_weights.shape}")

print(f"ivt_weights shape: {ivt_weights.shape}")
```

```python
node_labels_map = {
    (0,2): 'WLEH', (0,3): 'WLEH', (1,2): 'WLEH', (1,3): 'WLEH', (1,4): 'WLEH',
    (2,2): 'WLEH', (2,3): 'WLEH', (2,4): 'WLEH', (3,2): 'WLEH',
    (0,0): 'WHEL', (1,0): 'WHEL', (2,0): 'WHEL', (3,0): 'WHEL',
    (4,0): 'WHEL', (4,3): 'WHEL',
    (3,1): 'DH',   (4,1): 'DH',   (4,2): 'DH',   (4,4): 'DH',
    (0,1): 'DL',   (1,1): 'DL',   (3,3): 'DL',
    (0,4): 'NHSL', (2,1): 'NHSL', (3,4): 'NHSL',
}
lons = geo_daily.longitude.values - 360
lats = geo_daily.latitude.values
geo_abs_max = np.percentile(np.abs(geo_weights), 99)
pos_levels = np.linspace(0.1, geo_abs_max, 5)
neg_levels = np.linspace(-geo_abs_max, -0.1, 5)
ivt_abs_max = np.percentile(np.abs(ivt_weights), 99)
ivt_levels = np.linspace(-ivt_abs_max, ivt_abs_max, 21)
sgp_lon_min = float(precip_daily.longitude.values.min()) - 360
sgp_lon_max = float(precip_daily.longitude.values.max()) - 360
sgp_lat_min = float(precip_daily.latitude.values.min())
sgp_lat_max = float(precip_daily.latitude.values.max())
total_days = len(node_labels)

# precompute per-node precip anomaly (domain mean) — kept for reference
# node_precip_anom = {}
# for r in range(som_size):
#     for c in range(som_size):
#         mask = (node_labels == r * som_size + c)
#         node_precip_anom[(r,c)] = precip_anom_texas[mask].mean()

# precompute per-node precip net wet fraction
node_precip_frac = {}
for r in range(som_size):
    for c in range(som_size):
        mask = (node_labels == r * som_size + c)
        node_precip_map = precip_anom_np[mask].mean(axis=0)
        frac_above = (node_precip_map > 0).mean() * 100
        frac_below = (node_precip_map < 0).mean() * 100
        node_precip_frac[(r,c)] = frac_above - frac_below

# precompute per-node soil moisture anomaly (domain mean) — kept for reference
# node_sm_anom = {}
# for r in range(som_size):
#     for c in range(som_size):
#         mask = (node_labels == r * som_size + c)
#         node_sm_anom[(r,c)] = sm_anom_domain[mask].mean()

# precompute per-node soil moisture net wet fraction
node_sm_frac = {}
for r in range(som_size):
    for c in range(som_size):
        mask = (node_labels == r * som_size + c)
        node_sm_map = sm_anom_np[mask].mean(axis=0)
        frac_above = (node_sm_map > 0).mean() * 100
        frac_below = (node_sm_map < 0).mean() * 100
        node_sm_frac[(r,c)] = frac_above - frac_below

fig, axes = plt.subplots(som_size, som_size, figsize=(som_size*5, som_size*3),
                         subplot_kw={'projection': ccrs.PlateCarree()})

for r in range(som_size):
    for c in range(som_size):
        ax = axes[r, c]

        cf = ax.contourf(lons, lats, ivt_weights[r, c],
                         levels=ivt_levels, cmap='BrBG',
                         extend='both',
                         transform=ccrs.PlateCarree())
        cs_pos = ax.contour(lons, lats, geo_weights[r, c],
                            levels=pos_levels, colors='black', linewidths=1.2,
                            linestyles='solid', transform=ccrs.PlateCarree())
        cs_neg = ax.contour(lons, lats, geo_weights[r, c],
                            levels=neg_levels, colors='black', linewidths=1.2,
                            linestyles='dashed', transform=ccrs.PlateCarree())
        ax.clabel(cs_pos, inline=True, fontsize=14, fmt='%.1f')
        ax.clabel(cs_neg, inline=True, fontsize=14, fmt='%.1f')

        ax.set_extent([lons.min(), lons.max(), lats.min(), lats.max()],
                      crs=ccrs.PlateCarree())
        ax.add_feature(cfeature.COASTLINE, linewidth=0.3)
        ax.add_feature(cfeature.BORDERS,   linewidth=0.3)
        ax.add_feature(cfeature.STATES,    linewidth=0.3, edgecolor='black')

        ax.plot([sgp_lon_min, sgp_lon_max, sgp_lon_max, sgp_lon_min, sgp_lon_min],
                [sgp_lat_min, sgp_lat_min, sgp_lat_max, sgp_lat_max, sgp_lat_min],
                color='red', linewidth=1.2, transform=ccrs.PlateCarree())

        n = (node_labels == r * som_size + c).sum()
        freq = n / total_days * 100
        regime = node_labels_map.get((r, c), '')
        p_frac  = node_precip_frac[(r, c)]
        s_frac  = node_sm_frac[(r, c)]

        p_color  = '#185FA5' if p_frac >= 0 else '#D85A30'
        sm_color = '#185FA5' if s_frac >= 0 else '#D85A30'

        ax.set_title('')
        
        fontsize = 15

        ax.text(0.0,  1.01, f'({r},{c}) {regime}', transform=ax.transAxes,
                fontsize=fontsize, color='black', ha='left', va='bottom', fontweight='bold')
        ax.text(0.40, 1.01, f'P={p_frac:+.0f}%', transform=ax.transAxes,
                fontsize=fontsize, color=p_color, ha='center', va='bottom')
        ax.text(0.63, 1.01, f'SM={s_frac:+.0f}%', transform=ax.transAxes,
                fontsize=fontsize, color=sm_color, ha='center', va='bottom')
        ax.text(1.0,  1.01, f'f={freq:.1f}%', transform=ax.transAxes,
                fontsize=fontsize, color='black', ha='right', va='bottom')

plt.suptitle('SOM 5x5 — Z500 anomaly (contours) & IVT anomaly (shading) Warm season (Apr-Sep)',
             fontsize=26)
cbar_ax = fig.add_axes([0.2, -0.01, 0.6, 0.015])
cbar = plt.colorbar(cf, cax=cbar_ax, orientation='horizontal')
cbar.set_ticks([-1.25, -1, -0.75, -0.5, -0.25, 0, 0.25, 0.5, 0.75, 1, 1.25])
cbar.set_label('IVT anomaly (standardized)', fontsize=20)
cbar.ax.tick_params(labelsize=18, length=6, width=1.5)
plt.tight_layout()
plt.savefig('/home/annierosen16/som_5x5.png', dpi=120, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import cartopy.crs as ccrs
import cartopy.feature as cfeature

r, c = 0, 2
mask = (node_labels == r * som_size + c)

node_precip_map = precip_anom_np[mask].mean(axis=0)

print(f"Node ({r},{c}) — {mask.sum()} days")
print(f"Spatial anomaly range: {node_precip_map.min():.4f} to {node_precip_map.max():.4f}")
print(f"Fraction above zero: {(node_precip_map > 0).mean()*100:.1f}%")
print(f"Domain mean: {node_precip_map.mean():.4f}")

precip_lons = precip_daily.longitude.values
precip_lats = precip_daily.latitude.values

# days per year for this node
years_all = geo_daily.time.dt.year.values
unique_years = np.arange(years_all.min(), years_all.max() + 1)
days_per_year = np.array([(node_labels[years_all == y] == r * som_size + c).sum()
                           for y in unique_years])

levels = np.arange(-1.0, 1.1, 0.1)
cmap = plt.cm.BrBG
norm = mcolors.BoundaryNorm(levels, ncolors=cmap.N)

fig = plt.figure(figsize=(14, 6))
ax_map  = fig.add_axes([0.0,  0.0, 0.55, 1.0], projection=ccrs.PlateCarree())
ax_hist = fig.add_axes([0.62, 0.1, 0.35, 0.8])

# --- map ---
ax_map.set_extent([precip_lons.min()-1, precip_lons.max()+1,
                   precip_lats.min()-1, precip_lats.max()+1],
                  crs=ccrs.PlateCarree())
ax_map.add_feature(cfeature.COASTLINE, linewidth=0.8)
ax_map.add_feature(cfeature.BORDERS,   linewidth=0.8)
ax_map.add_feature(cfeature.STATES,    linewidth=0.5, edgecolor='black')

cf = ax_map.pcolormesh(precip_lons, precip_lats, node_precip_map,
                       cmap=cmap, norm=norm,
                       transform=ccrs.PlateCarree())

cbar = plt.colorbar(cf, ax=ax_map, orientation='horizontal', pad=0.05, shrink=0.8,
                    extend='both')
cbar.set_ticks(np.arange(-1.0, 1.1, 0.2))
cbar.set_label('Precip anomaly (mm/day)', fontsize=11)
cbar.ax.tick_params(labelsize=10)

gl = ax_map.gridlines(draw_labels=True, linewidth=0.4, color='#bbbbbb',
                      linestyle='--', alpha=0.7)
gl.top_labels   = False
gl.right_labels = False

ax_map.set_title(f'Node ({r},{c}) {node_labels_map.get((r,c), "")} — mean precip anomaly\n'
                 f'n={mask.sum()} days, {(node_precip_map > 0).mean()*100:.0f}% grid points above zero, '
                 f'domain mean={node_precip_map.mean():.3f} mm/day',
                 fontsize=11)

ax_hist.bar(unique_years, days_per_year, color='steelblue', width=0.7)
ax_hist.set_xlabel('Year', fontsize=11)
ax_hist.set_ylabel('Days assigned to node', fontsize=11)
ax_hist.set_title(f'Node ({r},{c}) — days per year', fontsize=11)
ax_hist.set_xticks(unique_years[::5])
ax_hist.tick_params(axis='x', labelrotation=45, labelsize=9)
ax_hist.tick_params(axis='y', labelsize=9)
ax_hist.axhline(days_per_year.mean(), color='red', linewidth=1.2,
                linestyle='--', label=f'Mean: {days_per_year.mean():.1f}')
ax_hist.legend(fontsize=9)

plt.savefig(f'/home/annierosen16/node_{r}{c}_precip_years.png', dpi=150, bbox_inches='tight')
plt.show()
```
