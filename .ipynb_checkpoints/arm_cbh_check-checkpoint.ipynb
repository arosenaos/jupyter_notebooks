```python
with open('/home/annierosen16/soundings_fig2&3.pdkl', 'rb') as file:
    
    soundings = pickle.load(file)
```

```python
## reading in cloud information

# Load file list
path = '/home/annierosen16/data/doe_arm_sgp/cloud_base_height/sgparsclbnd1clothC1.c1'

files = sorted(glob.glob(f"{path}/*.cdf"))

# Open individual datasets
datasets = [xr.open_dataset(f) for f in files]
```

```python
dfs = []

for ds in datasets:
    
    base_time = pd.to_datetime(ds['base_time'].values)
    
    time_offset = ds['time_offset'].values

    # Convert time_offset to seconds
    time_offset_sec = time_offset.astype('timedelta64[s]').astype(int)

    # Hourly mask
    hourly_mask = (time_offset_sec % 3600 == 0)

    # Compute full timestamps and subtract 6 hours
    times = base_time + pd.to_timedelta(time_offset_sec[hourly_mask], unit='s') - pd.Timedelta(hours=6)

    # Extract cloud base heights
    cloud_base = ds['CloudBaseBestEstimate'].values[hourly_mask]

    df = pd.DataFrame({
        'date': times.date,
        'hour': times.hour,
        'cloud_base_height': cloud_base
    })

    dfs.append(df)
    
cbh = pd.concat(dfs, ignore_index=True)
```

```python
cbh = cbh[cbh['date'].apply(lambda d: d.month in [5, 6, 7, 8, 9])]

daily_cbh = (cbh.groupby('date')['cloud_base_height']
       .apply(lambda x: x.to_numpy())
       .loc[lambda arrs: arrs.apply(len) == 24]
       .reset_index()
       .rename(columns={'cloud_base_height': 'cbh_array'})
)

daily_cbh.rename(columns={'cloud_base_height': 'cbh_array'}, inplace=True)
```

```python
# Calculate the average cbh_array for each hour (24 values)
avg_cbh = np.vstack(daily_cbh['cbh_array']).mean(axis=0)

# Plot the average cloud base height across hours
plt.figure(figsize=(10, 6))

plt.plot(range(1, 25), avg_cbh, marker='o', linestyle='-', color='b')

plt.title("Average Cloud Base Height Across Hours")

plt.xlabel("Hour of the Day")

plt.ylabel("Average Cloud Base Height")

plt.xticks(range(1, 25))

plt.grid(True)

plt.ylim(0, 4000)

plt.show()
```

```python
soundings_reset = soundings.reset_index()

# Ensure 'date' is datetime in both DataFrames
soundings_reset['date'] = pd.to_datetime(soundings_reset['date'])

daily_cbh['date'] = pd.to_datetime(daily_cbh['date'])

# Merge on 'date'
soundings_with_cbh = pd.merge(soundings_reset, daily_cbh[['date', 'cbh_array']], on='date', how='left')

# Optional: set 'date' as index again
soundings_with_cbh.set_index('date', inplace=True)
```

```python
# Function to replace values -3, -2, or -1 with NaN inside an array
def replace_invalid_cbh(arr):
    
    arr = np.where(np.isin(arr, [-3, -2, -1]), np.nan, arr)
    
    return arr

soundings_with_cbh['cbh_array'] = soundings_with_cbh['cbh_array'].apply(replace_invalid_cbh)
```

```python
soundings_with_cbh_clean = soundings_with_cbh[soundings_with_cbh['cbh_array'].notna()]
```

```python
dryapes = soundings_with_cbh[(soundings_with_cbh_clean['dry_coupling']==True) & (soundings_with_cbh['APE']==True)]

wetapes = soundings_with_cbh[(soundings_with_cbh_clean['wet_coupling']==True) & (soundings_with_cbh['APE']==True)]
```

```python
# cbh

dry_cbh_array = np.stack(dryapes['cbh_array'].values)

wet_cbh_array = np.stack(wetapes['cbh_array'].values)

dry_cbh_composite = np.nanmean(dry_cbh_array, axis=0) / 1000

wet_cbh_composite = np.nanmean(wet_cbh_array, axis=0) / 1000

dry_cbh_sem = np.nanstd(dry_cbh_array, axis=0) / np.sqrt(np.sum(~np.isnan(dry_cbh_array), axis=0)) / 1000

wet_cbh_sem = np.nanstd(wet_cbh_array, axis=0) / np.sqrt(np.sum(~np.isnan(wet_cbh_array), axis=0)) / 1000
```

```python
hours = list(range(24))

# Colors (muted palette)
dry_color = '#d95f02'        # Muted orange

wet_color = '#7570b3'        # Muted purple

# Create 1x2 subplot layout
fig, axs = plt.subplots(1, 2, figsize=(14, 5), sharex=True, sharey=True)

# --- Dry APE ---

axs[0].plot(hours, dry_cbh_composite, label='CBH', color=dry_color, lw=2, linestyle='--')

axs[0].fill_between(hours, 
                    dry_cbh_composite - dry_cbh_sem, 
                    dry_cbh_composite + dry_cbh_sem, 
                    color=dry_color, 
                    alpha=0.3, 
                    linestyle='--')

axs[0].set_title('Dry APE', fontsize=16)

axs[0].set_xlabel('Hour (LST)', fontsize=14)

axs[0].set_ylabel('Altitude (km)', fontsize=14)

axs[0].grid(True, linestyle='--', alpha=0.5)

axs[0].legend(fontsize=12)

axs[0].set_facecolor('#f9f9f9')

# --- Wet APE ---

axs[1].plot(hours, wet_cbh_composite, label='CBH', color=wet_color, lw=2, linestyle='--')

axs[1].fill_between(hours, 
                    wet_cbh_composite - wet_cbh_sem, 
                    wet_cbh_composite + wet_cbh_sem, 
                    color=wet_color, 
                    alpha=0.3, 
                    linestyle='--')

axs[1].set_title('Wet APE', fontsize=16)

axs[1].set_xlabel('Hour (LST)', fontsize=14)

axs[1].grid(True, linestyle='--', alpha=0.5)

axs[1].legend(fontsize=12)

axs[1].set_facecolor('#f9f9f9')

# Final layout adjustments
for ax in axs.flat:
    
    ax.set_xticks(hours)
    
    ax.tick_params(axis='both', labelsize=12)

fig.suptitle('Composite Cloud Base Height Diurnal Cycle', fontsize=18)

plt.tight_layout(rect=[0, 0, 1, 0.95])

plt.ylim(0, 8)

plt.show()
```
