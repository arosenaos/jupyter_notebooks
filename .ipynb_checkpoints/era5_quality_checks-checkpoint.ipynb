#### proof that 12pm q_obs - qfit and qfit are uncorrelated

```python
# load interpolated q for 12 LST

backup_dir = os.path.expanduser("~/backup_data")

arrays_file_path = os.path.join(backup_dir, "backup_arrays_for_buoyancy.npz")

metadata_file_path = os.path.join(backup_dir, "backup_metadata_for_buoyancy.npz")

arrays_data = np.load(arrays_file_path) 

meta_data = np.load(metadata_file_path)

lats = meta_data["lats"]

lons = meta_data["lons"]

interpolated_q = arrays_data['array2']

interpolated_qfit = arrays_data['array_4']

print("interpolated_pressure loaded")

print("interpolated_qfit loaded")
```

```python
time = np.arange(interpolated_q.shape[0])

level = np.arange(interpolated_q.shape[1])

lat = np.arange(interpolated_q.shape[2])

lon = np.arange(interpolated_q.shape[3])

qr = xr.DataArray(interpolated_q, dims=["time", "level", "lat", "lon"],
                 coords={"time": time, "level": level, "lat": lat, "lon": lon})

qfit = xr.DataArray(interpolated_qfit, dims=["time", "level", "lat", "lon"],
                    coords={"time": time, "level": level, "lat": lat, "lon": lon})

interpolated_altitudes = np.arange(0, 4000 + 20, 20)

qr = qr.assign_coords(level=interpolated_altitudes)

qfit = qfit.assign_coords(level=interpolated_altitudes)

mask = (interpolated_altitudes >= 2000) & (interpolated_altitudes <= 4000)

qr_subset = qr.sel(level=interpolated_altitudes[mask])

qfit_subset = qfit.sel(level=interpolated_altitudes[mask])

# residual

resid_subset = qr_subset - qfit_subset 

# compute correlation between residual and q fit at each gridpoint

corr = xr.corr(resid_subset, qfit_subset, dim="time")

corr_mean = corr.mean(dim="level")
```

```python
fig = plt.figure(figsize=(8, 5))

ax = plt.axes(projection=ccrs.PlateCarree())

ax.coastlines(resolution="50m", linewidth=1)

ax.add_feature(cfeature.BORDERS, linewidth=0.5)

ax.add_feature(cfeature.STATES, linewidth=0.4)

pcm = ax.pcolormesh(lons, lats, corr_mean, cmap="coolwarm", vmin=-1, vmax=1, transform=ccrs.PlateCarree())

cb = plt.colorbar(pcm, ax=ax, orientation="horizontal", pad=0.05, aspect=40, fraction=0.03)

cb.set_label("Pearson Correlation")

ax.set_title("Mean Correlation in LT layer between residual (qr - qfit) and qfit", fontsize=11)

plt.tight_layout()

plt.show()
```

#### how does lower tropospheric humidity in the morning influence BLT? 

```python
# add average surface pressure to q_sgp so that we can access LT layer properly

unique_sp = master[['latitude', 'longitude', 'average_sp']].drop_duplicates()

average_sp_2d = unique_sp.pivot(index='latitude', columns='longitude', values='average_sp')

average_sp_da = xr.DataArray(
    
    average_sp_2d.values,
    
    coords={
        
        'latitude': average_sp_2d.index.values,
        
        'longitude': average_sp_2d.columns.values}, dims=['latitude', 'longitude'])

q_sgp['average_sp'] = average_sp_da 

q_sgp['average_sp'] = q_sgp['average_sp'] / 100 
```

```python
p0 = q_sgp['average_sp']  # surface pressure at each gridpoint

p_100AGL = p0 - 100.0    # in hPa; 100 hPa above ground

p_300AGL = p0 - 300.0    # in hPa; 300 hPa above ground

# Step 2: round to nearest available level in q_sgp

levels = q_sgp['level'].values  # assumed in hPa

def nearest_level(p_array, levels):
    
    """Return nearest level in levels <= target p_array."""
    
    # broadcast p_array to levels for comparison
    
    p_array_exp = np.expand_dims(p_array, axis=0)  # shape (1, lat, lon)
    
    levels_exp = levels[:, np.newaxis, np.newaxis]  # shape (n_levels, 1, 1)
    
    # only consider levels lower in pressure (higher altitude)
    
    mask = levels_exp <= p_array_exp
    
    masked_levels = np.where(mask, levels_exp, np.nan)
    
    # pick the maximum level that satisfies mask (closest <= p)
    
    nearest = np.nanmax(masked_levels, axis=0)
    
    return nearest

nearest_100 = nearest_level(p_100AGL, levels)

nearest_300 = nearest_level(p_300AGL, levels)

# Step 3: select levels between nearest_100 and nearest_300

# convert nearest_100 and nearest_300 into xarray DataArrays for indexing

nearest_100_da = xr.DataArray(nearest_100, coords={'latitude': q_sgp['latitude'], 
                                                   
                                                   'longitude': q_sgp['longitude']}, 
                              
                              dims=['latitude', 'longitude'])

nearest_300_da = xr.DataArray(nearest_300, coords={'latitude': q_sgp['latitude'], 
                                                   
                                                   'longitude': q_sgp['longitude']}, 
                              
                              dims=['latitude', 'longitude'])

# Step 4: mask q_sgp to include only levels between nearest_100 and nearest_300

mask_layer = (q_sgp['level'] >= nearest_300_da) & (q_sgp['level'] <= nearest_100_da)

q_layer = q_sgp['q'].where(mask_layer, drop=True)

# Step 5: take average over the selected levels

q_avg_100_300AGL = q_layer.mean(dim='level')  # dims: (time, latitude, longitude)
```

```python
dryapes = master[master['dryape']==True].reset_index(drop=True)

dryapes['date'] = pd.to_datetime(dryapes['date'])
```

```python
# Initialize list to store q prior values

q_prior_list = []

for idx, row in dryapes.iterrows():
    
    lat = row['latitude']
    
    lon = row['longitude']
    
    date = pd.Timestamp(row['date'])
    
    # Select q at that gridpoint
    
    q_sel = q_avg_100_300AGL.sel(latitude=lat, longitude=lon, method='nearest')
    
    # Select times prior to 12 PM on that date
    
    day_end = date + pd.Timedelta(hours=11, minutes=59)
    
    q_pre_noon = q_sel.sel(time=slice(date, day_end))
    
    # Average over time
    
    q_mean = q_pre_noon.mean().compute().item()  # scalar
    
    print(lat, lon, date, q_mean)
    
    q_prior_list.append(q_mean)
    
# Add q_prior as a new column to dryapes

dryapes['ltq_prior'] = q_prior_list
```

```python
r, p = pearsonr(dryapes['blc_stnd_anom'], dryapes['ltq_prior'])

print(r, p)
```

```python
wetapes = master[master['wetape']==True].reset_index(drop=True)

wetapes['date'] = pd.to_datetime(wetapes['date'])

# Initialize list to store q prior values

q_prior_list = []

for idx, row in wetapes.iterrows():
    
    lat = row['latitude']
    
    lon = row['longitude']
    
    date = pd.Timestamp(row['date'])
    
    # Select q at that gridpoint
    
    q_sel = q_avg_100_300AGL.sel(latitude=lat, longitude=lon, method='nearest')
    
    # Select times prior to 12 PM on that date
    
    day_end = date + pd.Timedelta(hours=11, minutes=59)
    
    q_pre_noon = q_sel.sel(time=slice(date, day_end))
    
    # Average over time
    
    q_mean = q_pre_noon.mean().compute().item()  # scalar
    
    print(lat, lon, date, q_mean)
    
    q_prior_list.append(q_mean)
    
# Add q_prior as a new column to wetapes

wetapes['ltq_prior'] = q_prior_list
```

```python
r, p = pearsonr(wetapes['blc_stnd_anom'], wetapes['ltq_prior'])

print(r, p)
```
